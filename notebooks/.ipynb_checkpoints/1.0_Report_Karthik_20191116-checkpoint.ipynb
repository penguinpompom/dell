### 3.0 Data Preprocessing

```python
X = df.loc[:, ['Material1', 'Material2', 'Material3', 'Condition']]
X_drop_m1 = df.loc[:, ['Material2', 'Material3', 'Condition']]
y = df['Octane']
```

### 4.0 OLS Modelling

```python
model = sm.OLS(y, X).fit()
model.summary()
```

```python
model_drop_m1 = sm.OLS(y, X_drop_m1).fit()
model_drop_m1.summary()
```

__Insights:__ 
1. xxx
2. yyy
3. zzz

### 5.0 Model Diagnostics

```python
y_pred = model.predict(X)
y_pred_drop_m1 = model_drop_m1.predict(X_drop_m1)

residual = y - y_pred
residual_drop_m1 = y - y_pred_drop_m1
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2)

sns.scatterplot(residual, y, ax = axs[0][0])
sp.stats.probplot(residual, plot = axs[0][1], fit=True)

sp.stats.probplot(residual_drop_m1, plot = axs[1][1], fit=True)
sns.scatterplot(residual_drop_m1, y, ax = axs[1][0])
```
